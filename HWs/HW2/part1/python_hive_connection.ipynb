```python
from pyhive import hive  # or import hive or import trino
import pandas as pd
conn = hive.Connection(host="localhost", port=10000, username="root")
cursor = conn.cursor()
```

```python
cursor.execute("SELECT * FROM HW2.artist LIMIT 50")
data_list = []
for result in cursor.fetchall():
  data_list.append(result)

df = pd.DataFrame(data_list, columns=["mdid",
 "artist_mb",
 "artist_lastfm",
 "country_mb",
 "country_lastfm",
 "tags_mb",
 "tags_lastfm",
 "listeners_lastfm",
 "scrobbles_lastfm",
 "ambigous_artist"])
df
```
